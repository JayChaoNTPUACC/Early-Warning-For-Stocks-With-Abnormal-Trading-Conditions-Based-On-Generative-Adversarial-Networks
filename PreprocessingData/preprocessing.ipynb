```python
#import套件
import os
import pandas as pd
from datetime import datetime,timedelta
from sklearn.preprocessing import OneHotEncoder, RobustScaler
```

```python
#global variable
get_last_days = 20
```

```python
#載入資料
df_punishdata = pd.read_csv("get_stocks_result_1121119.csv")
```

```python
# 新增date_before_announced欄，其值均為0
df_punishdata = df_punishdata.assign(date_before_announced=0)
print(df_punishdata)
# 提取交易日期為公布日期前第幾個交易日的特徵
for i in range(len(df_punishdata)):
  df_punishdata['date_before_announced'][i] = (get_last_days - 1) - (i % get_last_days)
print(df_punishdata['date_before_announced'])
```

```python
# 對date_before_announced 和 MeasuringReasonCode 進行One-Hot-Encoding
df_punishdata = pd.get_dummies(df_punishdata , columns=['date_before_announced', 'MeasuringReasonCode'])
print(df_punishdata)
print(df_punishdata.columns)
```

```python
# 針對每一組資料除上開特徵外之數據進行Robust Scaling標準化

# 標準化之欄位
columns_to_scale = ['Trading_Volume','Trading_money','open','max','min','close','spread','Trading_turnover','MACD_macd','MACD_macdsignal','MACD_macdhist','RSI10','BOLL_Upper','BOLL_Middle','BOLL_Lower','CCI10','DX10','SMA5','SMA10', 'SMA20']

# 計算資料組數
data_groups_num = int(len(df_punishdata)/get_last_days)

j = 1
# 針對每一組資料進行Robust Scaling標準化
for i in range(data_groups_num):
  #初始化RobustScaler
  scaler = RobustScaler()
  #將每組資料設為temp_df
  temp_df = df_punishdata.loc[(get_last_days*i):(get_last_days*i+get_last_days-1)]
  #對該組數據針對指定標準化之欄位進行 Robust Scaling
  temp_df[columns_to_scale] = scaler.fit_transform(temp_df[columns_to_scale])
  #print(temp_df)
  if j == 1:
    new_df_alldatas = temp_df
  else:
    new_df_alldatas = pd.concat([new_df_alldatas, temp_df], ignore_index=True)
  j += 1
```
